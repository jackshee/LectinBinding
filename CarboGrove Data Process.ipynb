```python
df_human_lectins = pd.read_csv('data/human_lectins.tsv', delimiter='\t')
df_human_lectins
```

```python
gene_names = df_human_lectins['HGNC gene name'].tolist()
len(gene_names)
```

```python
from Bio import Entrez, SeqIO

Entrez.email = "jshee@student.unimelb.edu.au"  # Always use your own email address here.

def search_gene(hgnc_name):
    handle = Entrez.esearch(db="gene", term=f"{hgnc_name}[sym] AND Homo sapiens[orgn]")
    record = Entrez.read(handle)
    handle.close()
    gene_ids = record["IdList"]
    return gene_ids

# Example usage:
gene_ids = [ search_gene(HGNC_gene_name) for HGNC_gene_name in gene_names ]

df_human_lectins['NCBI gene ID'] = df_human_lectins['HGNC gene name'].apply(lambda x: search_gene(x))

df_human_lectins
```
